# Import Packages

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Load Data

```python
dataset = pd.read_csv('Poly_dataSet.csv')
X = dataset.iloc[:,0:1].values
y = dataset.iloc[:,1].values
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=0)
dtr.fit(X,y)
```

# Plotting 

```python
plt.scatter(X,y,color='red')
plt.plot(X,dtr.predict(X),color='blue')
plt.title("DTR Salary vs Expirenece")
plt.xlabel("Position of Employee")
plt.ylabel("Salaries of Employees")
plt.show()
```

```python
x_grid = np.arange(min(X),max(X),0.1)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(X,y,color='red')
plt.plot(x_grid,dtr.predict(x_grid),color='blue')
plt.title("Truth or Bluff (DT Regresssion)")
plt.xlabel("Position of Employee")
plt.ylabel("Salaries of Employees")
plt.show()
```
